implementing sentence transformer

```python

%pip install -U sentence-transformers

from sentence_transformers import SentenceTransformer
```

```python
try:
    model = SentenceTransformer("all-MiniLM-L6-v2")
except Exception as e:
    print("Failed to load 'all-MiniLM-L6-v2':", e)
    try:
        # fallback: explicit repo id and force CPU (useful if CUDA/torch issues)
        model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2", device="cpu")
        print("Loaded fallback model on CPU")
    except Exception as e2:
        print("Fallback load failed:", e2)
        raise
```

```python
print(type(model))
```
